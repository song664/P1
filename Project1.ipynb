```python
#2
df = df.dropna(subset=['BOROUGH'])

populations = {
    'Bronx': 1446788,
    'Brooklyn': 2648452,
    'Manhattan': 1638281,
    'Queens': 2330295,
    'Staten Island': 487155,
}

borough_counts = df['BOROUGH'].value_counts()

crashes_per_100k = {}
for borough, crashes in borough_counts.items():
    corrected_borough = borough.title()
    
    if corrected_borough not in populations:
        print(f"Skipping unknown borough: {borough}")
        continue
    
    population = populations[corrected_borough]
    crashes_per_100k[corrected_borough] = (crashes / population) * 100000


if not crashes_per_100k:
    print("No data in crashes_per_100k. Please check the borough names.")
else:
   
    max_borough = max(crashes_per_100k, key=crashes_per_100k.get)
    print(f"{max_borough}, with {crashes_per_100k[max_borough]:.5f} crashes per 100k population")
```

```python
import pandas as pd


df = pd.read_pickle(r"shared/Motor_Vehicle_Collisions_-_Crashes.pkl")

```

```python
#3
df = df[df['CONTRIBUTING FACTOR VEHICLE 1'].notna() & (df['CONTRIBUTING FACTOR VEHICLE 1'] != 'Unspecified')]

factor_counts = df['CONTRIBUTING FACTOR VEHICLE 1'].value_counts()

leading_cause = factor_counts.idxmax()
proportion = (factor_counts.max() / factor_counts.sum()) * 100

print(f"{leading_cause}, {proportion:.1f}%")
```

```python
4 # Sum the counts of the top 3 causes and divide by the total count of specified crashes
top_3_proportion = factor_counts.head(3).sum() / factor_counts.sum() * 100

print(f"{top_3_proportion:.1f}%")
```
